```python
%matplotlib inline
from PIL import Image
import matplotlib.pyplot as plt
import math
```

# 穴が開くように回転するプログラム

```python
src_img = Image.open("4.2.03.tiff")
plt.imshow(src_img)
width, height = src_img.size
px = width / 2
py = height / 2

dst_img = Image.new("RGB", (width * 2, height * 2))
dwidth, dheight = dst_img.size
cx = dwidth / 2
cy = dheight / 2

th = math.radians(45)
for y in range(height):
    for x in range(width):
        r, g, b = src_img.getpixel((x, y))
        dx = int((x - px) * math.cos(th) - (y - py) * math.sin(th) + cx)
        dy = int((x - px) * math.sin(th) + (y - py) * math.cos(th) + cy)
        
        dst_img.putpixel((dx, dy), (r, g, b))

plt.figure(figsize=(8, 8))
plt.imshow(dst_img)
plt.show()
```

# 穴があかないように回転するプログラム

```python
src_img = Image.open("4.2.03.tiff")
plt.imshow(src_img)
width, height = src_img.size
px = width / 2
py = height / 2

dst_img = Image.new("RGB", (width * 2, height * 2))
dwidth, dheight = dst_img.size
cx = dwidth / 2
cy = dheight / 2

th = math.radians(45)
for y in range(dheight):
    for x in range(dwidth):
        x0 = int((x - cx) * math.cos(-th) - (y - cy) * math.sin(-th) + cx - px)
        y0 = int((x - cx) * math.sin(-th) + (y - cy) * math.cos(-th) + cy -py)
        
        if ((x0 >= 0 and x0 < width) and (y0 >= 0 and y0 < height)):
            r, g, b = src_img.getpixel((x0, y0))
            dst_img.putpixel((x, y), (r, g, b))

plt.figure(figsize=(8, 8))
plt.imshow(dst_img)
plt.show()
```
